```python
## Import Ticker class from yahooquery library 
# To access data from yahooquery, an argument (Company ticker symbol) is passed to the Ticker class 

from yahooquery import Ticker
import pandas as pd
```

```python
## Create instance of Ticker class & pass argument (Company ticker symbol)

WES_ASX = Ticker(['WES.AX'])
```

```python
# Select the Balance Sheet from the Ticker class, this will return a DataFrame

balanceSheet = WES_ASX.balance_sheet()

# Print top 5 results 

balanceSheet.head()
```

```python
### Liquidity Analysis 

# Extract each required column for liquidity analysis from the balanceSheet object 


# Reporting Period

reportingPeriod = balanceSheet['asOfDate']

# Total Assets 

totalAssets = balanceSheet['TotalAssets']

# Current Assets

currentAssets = balanceSheet['CurrentAssets']

# Non Current Assets 

nonCurrentAssets = balanceSheet["TotalNonCurrentAssets"]

# Total Liabilities 

totalLiabilities = balanceSheet['TotalLiabilitiesNetMinorityInterest']

# Non Current Liabilities 

nonCurrentLiabilities = balanceSheet['TotalNonCurrentLiabilitiesNetMinorityInterest']

# # Current Liabilities 

currentLiabilities = totalLiabilities - nonCurrentLiabilities 

# Inventory 

inventory = balanceSheet['Inventory']
```

```python
## Create new DataFrame containing each required column with pd.concat method, which concentrates DataFrame objects
# Add relevant keys for each column 



liquidityAnalysis = pd.concat([reportingPeriod, totalAssets, currentAssets, nonCurrentAssets,
                               totalLiabilities, currentLiabilities, nonCurrentLiabilities], 
                         axis=1, keys=['Reporting Period', 'Total Assets', 
                                       'Current Assets', 'Non-Current Assets',
                                       'Total Liabilities', 'Current Liabilities', 
                                       'Non-Current Liabilities'])
```

```python
# Print results

liquidityAnalysis
```

```python
## Debt-to-Assets Ratio 

liquidityAnalysis['Debt to Assets Ratio'] = liquidityAnalysis['Total Assets'] / liquidityAnalysis['Total Liabilities']
```

```python
## Current Ratio 

liquidityAnalysis['Current Ratio'] = liquidityAnalysis['Current Assets'] / liquidityAnalysis['Current Liabilities']
```

```python
## Acid Test 

liquidityAnalysis['Acid Test'] = (liquidityAnalysis['Current Assets'] - balanceSheet['Inventory']) / liquidityAnalysis['Current Liabilities']
```

```python
# Print results

liquidityAnalysis
```
